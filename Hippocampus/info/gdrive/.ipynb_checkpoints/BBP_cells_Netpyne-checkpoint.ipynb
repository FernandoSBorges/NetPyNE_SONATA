## Netpyne

```python
%matplotlib inline
```

```python
from netpyne import specs, sim   

# Network parameters
netParams = specs.NetParams()  # object of class NetParams to store the network parameters
```

```python
for gid in [14444]:
    MorphoName = nodesinfo['morphology'][gid] + '.swc'
    hocName = nodesinfo['model_template'][gid][4:]  
    if cellName == nodesinfo['mtype'][gid] + '_' + nodesinfo['etype'][gid] + '_' + nodesinfo['region'][gid][:3]:
        cellName = nodesinfo['mtype'][gid] + '_' + nodesinfo['etype'][gid] + '_' + nodesinfo['region'][gid][:3] + '_2'
    else:  
        cellName = nodesinfo['mtype'][gid] + '_' + nodesinfo['etype'][gid] + '_' + nodesinfo['region'][gid][:3]
    
    cellRule = netParams.importCellParams(label=cellName, somaAtOrigin=False,
        conds={'cellType': cellName, 'cellModel': 'HH_full'},
        fileName='cellwrapper.py',
        cellName='loadCell',
        cellInstance = True,
        cellArgs={'hocName': hocName, 'MorphoName': MorphoName})
    netParams.renameCellParamsSec(label=cellName, oldSec='soma_0', newSec='soma')
```

```python
# netParams.cellParams[cellName]
netParams.cellParams.keys()
```

```python
netParams.popParams['SP_PC_cACpyr_mc2_2'] = {'cellType': 'SP_PC_cACpyr_mc2_2', 'numCells': 1, 'cellModel': 'HH_full'}
```

```python
# netParams.synMechParams['SP_PC:SP_PC'] = {'mod': 'DetAMPANMDA'}

netParams.synMechParams['SP_PC:SP_PC'] = {'mod': 'DetAMPANMDA',
                                          'Dep': 671.0,
                                          'Fac': 17.0,
                                          'NMDA_ratio': 1.22,
                                          'Use': 0.5,
                                          'tau_d_AMPA': 3.0,
                                          'tau_d_NMDA': 148.5}

# netParams.synMechParams['PC-PC'] = {'mod': 'DetAMPANMDA','tau_d_AMPA': 3.0,'tau_d_NMDA': 148.5,'Use': 0.5,'Dep': 671.0,'Fac': 17.0, 'NMDA_ratio': 1.22}
                                          
netParams.stimSourceParams['Input'] = {'type': 'NetStim', 'rate':20, 'start': 600, 'number': 8, 'noise': 0.0}
# netParams.stimSourceParams['Input'] = {'type': 'NetStim', 'interval': [0,50,50,50,50,50,50,50,550], 'number': 9, 'start': 600, 'noise': 0.0}

netParams.stimTargetParams['Input->SP_PC_cACpyr'] = {'source': 'Input','synMech': 'SP_PC:SP_PC', 'weight': 0.6,'delay': 0.5, 'conds': {'pop':'SP_PC_cACpyr_mc2_2'}}
```

```python
netParams.synMechParams
```

```python
## cfg  
cfg = specs.SimConfig()					            # object of class SimConfig to store simulation configuration
cfg.duration = 1300.0 						            # Duration of the simulation, in ms
cfg.dt = 0.01								                # Internal integration timestep to use
cfg.verbose = False							                # Show detailed messages 
cfg.recordTraces = {'V_soma':{'sec':'soma','loc':0.5,'var':'v'}}  # Dict with traces to record
cfg.recordStep = 0.01 			
cfg.filename = 'model_output'  			# Set file output name
cfg.saveJson = False
cfg.analysis['plotTraces'] = {'include': [0], 'timeRange': [500,1300], 'saveFig': True, 'showFig': True, 'figSize':(12,4)} # Plot recorded traces for this list of cells
# cfg.analysis['plotShape'] = {'includePre': [0],'includePre': [0], 'saveFig': True, 'showFig': True, 'figSize':(12,12)}
cfg.hParams['celsius'] = 34.0
```

```python
sim.initialize(
    simConfig = cfg, 	
    netParams = netParams)  				# create network object and set cfg and net params
sim.net.createPops()               			# instantiate network populations
```

```python
sim.net.createCells()              			# instantiate network cells based on defined populations
```

```python
sim.net.connectCells()            			# create connections between cells based on params
sim.net.addStims() 							# add network stimulation
```

```python
sim.net.params.synMechParams
```

```python
sim.setupRecording()              			# setup variables to record for each cell (spikes, V traces, etc)
sim.runSim()                      			# run parallel Neuron simulation  
```

```python
sim.gatherData()                  			# gather spiking data and cell info from each node
sim.saveData()                    			# save params, cell info and sim output to file (pickle,mat,txt,etc)#
sim.analysis.plotData()         			# plot spike raster etc
```

```python
sim.analysis.plotTraces(oneFigPer='trace', overlay=0, timeRange=[500,1300])
```
