```python
import matplotlib.pyplot as plt
```

```python
f_new = h5py.File('/home/fernando/Documents/SCx_model/O1_data_physiology/S1nonbarrel_neurons/nodes_new.h5', 'r')
```

```python
x_new = []
y_new = []
z_new = []
hexagon = []

for gid in list(f_new['nodes_new']['axis1']):    
    x_new.append(f_new['nodes_new']['block0_values'][gid][0])
    y_new.append(f_new['nodes_new']['block0_values'][gid][1])
    z_new.append(f_new['nodes_new']['block0_values'][gid][2])
    hexagon.append(f_new['nodes_new']['block1_values'][gid][0])
```

```python
# plot the results
fig=plt.figure(figsize=(18,6))

plt.subplot(1,3,1)
plt.scatter(x_new,y_new, lw=0.05, s=18,
            c=hexagon, cmap=plt.cm.get_cmap('gnuplot', 12))
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)


plt.subplot(1,3,2)
plt.scatter(x_new,z_new, lw=0.05, s=18,
            c=hexagon, cmap=plt.cm.get_cmap('gnuplot', 12))
plt.xlabel('x', fontsize=12)
plt.ylabel('z', fontsize=12)


plt.subplot(1,3,3)
plt.scatter(y_new,z_new, lw=0.05, s=18,
            c=hexagon, cmap=plt.cm.get_cmap('gnuplot', 12))
plt.xlabel('y', fontsize=12)
plt.ylabel('z', fontsize=12)
```
